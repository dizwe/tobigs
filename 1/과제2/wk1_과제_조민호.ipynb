#### Auction_Class나 Bid Class, addr_do, Apartment_usage는 실제로 금액에 영향이 있는지를 target encoding으로 변환해볼 수도 있겠다.
##### 질문) 그런데 명목형에 번호 붙여서 Ordinal 처럼 사용하면 안되나여..? -> 순서를 둔다는 건 s M L 처럼 숫자가 올라갈수록 어떤 값이 바뀐다는걸 보ㅕ주기 위함이다. 때문에 순서 없는건 안하는게 좋음

```python
# 7천만원 정도의 차이가 있으므로 target_encoding 해볼 만 하다.
data.groupby('Auction_class')['Hammer_price'].mean()
```

```python
## 질문) target encoding 하면 강제인지 임의인지 파악하기가 힘든데 보통 replace하나요?
## 질문) 이걸 한줄로 할 방법은 없나요? -> 알려주심 dict map
groupby_table = data.groupby('Auction_class')['Hammer_price'].mean()
data.loc[data['Auction_class']=='강제','Auction_class_enc'] = groupby_table['강제']
data.loc[data['Auction_class']=='임의','Auction_class_enc'] = groupby_table['임의']
```

```python
# 유의미한 차이가 있으므로 target encoding 사용하는게 좋을 듯
data.groupby('Bid_class')['Hammer_price'].mean()
```

```python
# 아예 다른 class 로 만들어보기
groupby_table_bid = data.groupby('Bid_class')['Hammer_price'].mean()
data.loc[data['Bid_class']=='개별','Bid_clas_enc'] = groupby_table_bid['개별']
data.loc[data['Bid_class']=='일괄','Bid_clas_enc'] = groupby_table_bid['일괄']
data.loc[data['Bid_class']=='일반','Bid_clas_enc'] = groupby_table_bid['일반']
```

```python
data['Bid_clas_enc'].head()
```

```python
# 영향력이 커보이므로 target encoding 사용하는게 좋을 듯
data.groupby('addr_do')['Hammer_price'].mean()
```

```python
groupby_table_do = data.groupby('addr_do')['Hammer_price'].mean()
data.loc[data['addr_do']=='서울','addr_do_enc'] = groupby_table_do['서울']
data.loc[data['addr_do']=='부산','addr_do_enc'] = groupby_table_do['부산']
```

```python
data['addr_do_enc'].head()
```

```python
# 차이가 없어보이므로 target_encoding은 별로일듯
data.groupby('Apartment_usage')['Hammer_price'].mean()
```
